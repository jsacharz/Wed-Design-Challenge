```python
import pandas as pd
import random
```

```python
file = pd.read_csv("data_homework.csv")
```

```python
reader = file
```

```python
reader.head()
```

```python
proteomics_df = pd.DataFrame(reader)
proteomics_df
```

```python
proteomics_new_df = proteomics_df[[
       'Protein ID', 'Log2 enrichment','T-test Significant',
       'Test statistic', 'Sample_1_Control', 'Sample_2_Control',
       'Sample_3_Control', 'Patient_1', 'Patient_2', 'Patient_3']]
```

```python
proteomics1_new_df = proteomics_new_df.sample(100)
proteomics1_new_df
```

```python
proteomics1_new_df['Log2 enrichment'] = proteomics1_new_df['Log2 enrichment'].map('{:,.2f}'.format)
proteomics1_new_df['Test statistic'] = proteomics1_new_df['Test statistic'].map('{:,.2f}'.format)

```

```python
proteomics2_new_df = proteomics1_new_df.reset_index(drop=True)
proteomics2_new_df
```

```python
proteomics_db_html = proteomics2_new_df.to_html(classes=["table-bordered", "table-striped", "table-hover"])
```

```python
table_html = proteomics_db_html.replace('\n','')
table_html
```
